```python
#### Dependencies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from Functions.Preprocess import tracks_to_df
```

```python
#### Parameters
max_ev = 50000
sort = False
PT_cut = [-1000, 1000]
rinv1_path = "Data/from_cluster/30.9.20/dark_test2.10000.0.5.root.GetTracks.txt"
rinv2_path = "Data/from_cluster/20.9.20/dark1.root.GetTracks.txt"
rinv3_path = "Data/from_cluster/dark1.root.GetTracks.txt"
n_constits=1
```

```python
%%time
rinv1 = tracks_to_df(rinv1_path, label=1, max_ev=max_ev, n_constits=n_constits, trunc=False, PT_cut=PT_cut, sort=sort) 
rinv2 = tracks_to_df(rinv2_path, label=1, max_ev=max_ev, n_constits=n_constits, trunc=False, PT_cut=PT_cut, sort=sort)
rinv3 = tracks_to_df(rinv3_path, label=1, max_ev=max_ev, n_constits=n_constits, trunc=False, PT_cut=PT_cut, sort=sort)
rinv1["track_mult"] = rinv1.track_PT.map(lambda x: len(x))
rinv2["track_mult"] = rinv2.track_PT.map(lambda x: len(x))
rinv3["track_mult"] = rinv3.track_PT.map(lambda x: len(x))
```

```python
print(len(rinv1), len(rinv2), len(rinv3))
```

```python
rinv1.jet_PT.hist(range=[0,400], bins=100, density=True, histtype='step', hatch='x', color='red')
rinv2.jet_PT.hist(range=[0,400], bins=100, density=True, histtype='step', hatch='x', color='black')
rinv3.jet_PT.hist(range=[0,400], bins=100, density=True, histtype='step', hatch='x', color='blue')
plt.gcf().set_size_inches(10, 10)
```
